## 迴圈(loop)


### range

```python
a = 2
b = 10
c = 3
```
相當於下列兩種方式
```python
a = 2; b = 10; c = 3
```
```python
a, b, c = 2, 10, 3
```

範例

````python
a = 2
b = 10
c = 3
````

````python
range(a) # 0, 1
````

````python
range(a, b) # 2, 3, ..., 9
````

````python
range(a, b, c) # 2, 5, 8
````

### for loop

````python
for i in range(1, 6):
  print(i)
````

### While loop

````python
i = 0
while (i < 3):
  print(i)
  i = i + 1 # 相當於 i += 1
````

for 寫法

````python
for i in range(3):
  print(i)
````

### break、continue、pass

````python
for i in range(4):
  if i == 2:
    pass
    print('QQ')
  else:
    print(i)
````

````python
for i in range(4):
  if i == 2:
    continue
    print('QQ')
  else:
    print(i)
````

````python
for i in range(4):
  if i == 2:
    break
    print('QQ')
  else:
    print(i)
````
