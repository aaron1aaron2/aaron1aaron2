### Optimizer
- **梯度下降法 (Gradient Descent)**:
    - **隨機梯度下降（SGD）**: 目標是最小化（或最大化）一個損失函數，這個函數衡量模型預測的輸出和實際目標的差異。梯度是損失函數對於模型參數的偏導數，它指示著損失函數在參數空間中的變化方向。
    - **Momentum**: 是一種在梯度下降法中用於加速收斂的技巧，它有助於克服收斂過程中的震盪和不必要的振動。Momentum的概念是基於物理中動量的概念，它引入了一種「動量」項來保持更新的方向，有助於突破局部極小值。
    > *這類方法，學習率（learning rate）是需要手動調整的超參數，因為過大的學習率可能導致優化過快，無法收斂；而過小的學習率可能導致收斂速度過慢，或者在局部極小值處停滯。*

  

- **自適應法 (Adaptive optimization methods)**
    - **Adagrad**：基於每個參數的歷史梯度來調整學習率。
    - **RMSprop**：是Adagrad的變種，使用遞動平均數來調整梯度的平方。
    - 👑 **Adam** ：結合了動量（Momentum）和自適應學習率的特點，同時使用梯度和梯度平方的移動平均數。
    > *儘管這些優化器通常能夠自動調整學習率，但在實際應用中，有時還是需要根據具體的問題和數據集進行一些超參數的調整。*

## How to reduce training time?



### Hardware Acceleration
- **GPU/TPU Usage**: Leverage graphics processing units (GPUs) or tensor processing units (TPUs) to accelerate training, especially for large models and datasets.
- **Reduce IO frequency**:
    Reducing input/output (IO) frequency during model training is crucial for improving training efficiency, especially when dealing with large datasets. High IO frequency can lead to significant overhead and slow down the training process.
    - Reduce Logging Frequency
    - Use periodic model checkpointing
    - Use larger batch sizes
    - Use Efficient Data Formats: For example, binary formats like HDF5 or TFRecord can be more efficient than plain text formats.



### Use faster algorithms
- 訓練速度: 1 epoch 跑多久?
    - 減少模型參數量
    - 使用較快的模型，像是同系列的 sequence model，GRU 通常會比 Lstm 快(大多數情況下他們效果都會優於 RNN)。

    - 根據任務選擇合適的模型，影像方面 CNN 會優於一般的 DNN。

### Hyperparameter
- Learning Rate: 越大收斂越快，但是太大可能會無法收斂。
- Epoch: 根據目 train、val loss 的變化曲線調整訓練的總次數，可以有效減少單次實驗的時間。也可以使用上面提到的 patience 方式，讓模型效果沒有再提昇時不用跑滿設定的 epoch 次數。
- Batch size: 越大收斂越快，且可以減少跑完一個 epoch 需要的迭代次數。缺點是越大會需要更多 GPU 專屬 Ram，且最後模型效果不一定會較好。

```python
def plot_data(img_arr:np.ndarray, label_arr:np.ndarray, target=None, rows = 2, columns = 5):
    # 選擇要顯示的 10 張圖片
    imgs = []
    labels = []
    for i in range(0, 10):
        if target:
            tar_arr = np.where(y_train==target)[0]
        else:
            tar_arr = np.where(y_train==i)[0]
        idx = np.random.choice(tar_arr)
        imgs.append(x_train[idx])
        labels.append(y_train[idx])


    # 畫圖
    plt.clf()
    fig, axes = plt.subplots(nrows=rows, ncols=columns, figsize=(columns*3, rows*3))

    axes = axes.ravel()

    for i in range(rows*columns):
        axes[i].imshow(imgs[i], cmap='gray')
        axes[i].set_title(f"Label: {labels[i]}")
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()
```

建立 validation set

```python
import torch

# 計算 val 和新的 train 大小
val_size = 10000 # 和 test 一樣數量
new_train_size  = len(x_train) - val_size

# 使用 index 切 (因為我們現在的資料已經把 data 和 label 分開了)
new_train_idx, val_idx = torch.utils.data.random_split(range(len(x_train)), [new_train_size, val_size])

# 切出新的 train & val
x_train_new = x_train[np.array(new_train_idx)]
y_train_new = y_train[np.array(new_train_idx)]
x_val = x_train[np.array(val_idx)]
y_val = y_train[np.array(val_idx)]
```

### Step2: 建立 dataloader

```python
class CustomDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = {
            'data': torch.Tensor(self.data[idx]),
            'label': torch.Tensor([self.labels[idx]])
        }
        return sample
```

```python
train_data = CustomDataset(x_train_new, y_train_new)
train_dataloader = DataLoader(train_data, batch_size=batch_size, shuffle=True)

val_data = CustomDataset(x_val, y_val)
val_dataloader = DataLoader(val_data, batch_size=batch_size, shuffle=True)

test_data = CustomDataset(x_val, y_val)
test_dataloader = DataLoader(val_data, batch_size=batch_size, shuffle=True)
```

### Step3: 建立 CNN model

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)

        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = x.unsqueeze(1) # [100, 28, 28] -> 100, 1, 28, 28]
        x = self.pool1(self.relu1(self.conv1(x))) # torch.Size([64, 14, 14])

        x = self.pool2(self.relu2(self.conv2(x))) # torch.Size([32, 7, 7])
        x = self.flatten(x) # torch.Size([32, 49])
        x = self.relu3(self.fc1(x))
        x = self.fc2(x)

        return x
```

```python
# Instantiate the model
model = CNN()

# Move the model to the GPU if available
model.to(device)
summary(model, (28, 28), batch_size=batch_size)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
losses = []
accuracies = []
val_losses = []
val_accuracies = []
# Train the model
for epoch in range(num_epochs):
    for i in train_dataloader:
        # 這邊 i 根據上面 CustomDataset 定義為一個 dict，包含 data、label
        images = i['data'].to(device)
        labels = i['label'].to(device).long().squeeze()
        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
    optimizer.step()

        _, predicted = torch.max(outputs.data, 1)
    acc = (predicted == labels).sum().item() / labels.size(0)
    accuracies.append(acc)
    losses.append(loss.item())

    # Evaluate the model on the validation set
    val_loss = 0.0
    val_acc = 0.0
    with torch.no_grad():
        for i in val_dataloader:
            images = i['data'].to(device)
            labels = i['label'].to(device).long().squeeze()
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            _, predicted = torch.max(outputs.data, 1)
        total = labels.size(0)
        correct = (predicted == labels).sum().item()
        val_acc += correct / total
        val_accuracies.append(acc)
        val_losses.append(loss.item())


    print('Epoch [{}/{}],Loss:{:.4f},Validation Loss:{:.4f},Accuracy:{:.2f},Validation Accuracy:{:.2f}'.format(
        epoch+1, num_epochs, loss.item(), val_loss, acc ,val_acc))
```

# 補充: difference between Loss function and Metrics

||Loss function|Metrics|
|-|-|-|
|Focus|guiding the model’s training process by quantifying<br /> how well it performs on individual training examples|summarizing the model’s overall performance on <br />validation or test sets|
|Differentiability|must be differentiable for optimization purposes <br />during training| _ |
|Training vs. Generalization|is crucial for the training phase, where the model <br />learns from the training data|come into play during the post-training phase to<br /> measure the model’s ability to generalize to new <br />data.|
|Optimization Objective|serves as the optimization objective during training,<br /> determining how the model adjusts its parameters|provide a means for comparing models and <br />assessing their effectiveness in real-world <br />scenarios|
