##  EDA 流程

### step 1 - 檢查與了解資料
- <font color="#48C9B0">資料來源與蒐集方式</font>
    - 目的是確保資料的可靠性，和預先知道使用該資料時會面臨到的問題。
    - EX: 由道路 senser 蒐集的車流資料，我們可以知道會有多個時間序列資料，所以我們要先了解有多少 senser? 多久會記錄一筆資訊? 這樣才能幫助我們做後續分析與使用。 此外，我們有可以預先知道正常情況下車流不可能為負值，所以當資料為負的就代表有問題。資料中出現 0 ，我們需要去確認是該 senser 失效還是只是那段時間沒有車經過而已。
- <font color="#48C9B0">資料本身</font>:
    - 了解每個欄位的意義、資料類別。
    - 有時候資料太複雜，或是品質不佳還要評估是否要使用該資料或是欄位，確保資料品質。
    - 要去思考資料本身是否符合需求 or 能否解決當前問題。資料量不足、資料不夠詳細。
    - 數據洩漏(Data Leakage)的問題




### step 2 - 基本資料處理
- <font color="#48C9B0">必要動作</font>: 異常值(Outliers)、缺失值(Missing values)、重複值(Duplicate values)、確保資料類型。目的是確保後續分析可以順利進行和結果的有效性。
- <font color="#48C9B0">視情況使用</font>: 標準化(Standardization)、正規化(Normalization)、取 log、非結構化資料的處理。這邊的目的只是輔助下一步的分析，會保留原始資料。
    - 標準化: z-score → 資料為高斯分布、存在 Outliers 時
    - 正規化: Min-Max scaling → 資料非高斯分布、對 Outliers 敏感、標準差較小時
    - 取 log →長尾型偏斜分布的資料





### step 3 - 統計分析與視覺化
- <font color="#48C9B0">單變數分析</font>
    - 數值、有序類別: 資料的分布(<font color="#EA9F13 ">Histogram, 直方圖</font>)、基本統計值(<font color="#EA9F13 ">Boxplot, 箱形圖</font>)。
    - 無序類別: 比例(<font color="#EA9F13 ">Pie chart, 圓餅圖</font>)、數量排名(<font color="#EA9F13 ">Bar chart, 長條圖</font>)。
- <font color="#48C9B0">多變數分析</font>
    - 雙變數
        - 數值 & 類別 - 視覺化分析(<font color="#EA9F13 ">Grouped boxplot, 分組箱形圖</font>)、T 檢定(<font color="#EA9F13 ">T-test</font>)
         - <i><font color="gray"> p.s T 檢定主要常用的是檢驗兩組之間是否有均值的差異，所以一次只能筆比較兩組。 </font></i>
        - 數值 & 數值 -  視覺化分析(<font color="#EA9F13 ">Scatter plot, 散佈圖</font>)、相關係數(<font color="#EA9F13 ">Correlation coefficient</font>)
        - 類別 & 類別 - 視覺化分析(<font color="#EA9F13 ">Grouped bar chart, 分組長條圖</font>)、卡方檢定 (<font color="#EA9F13 ">Chi-Squared Test</font>)
    - 多變數 >= 2
        - 多數值 → 多數值: 視覺化分析(<font color="#EA9F13 ">Heatmap, 熱圖</font>)、共線性分析
        - 多數值 → 數值: 回歸分析(<font color="#EA9F13 ">Multiple Regression</font>)
        - 多類別 → 數值:  變異數分析(<font color="#EA9F13 ">ANOVA</font>)
